## ガウス過程回帰によるヤコビ行列推定
$$
\begin{align}
\frac{dx}{dt} = f(x(t))
\end{align}
$$
という力学系について、任意の状態$x(t)$の$\Delta t$後の状態を与える関数が存在するとする。
$$x(t + \Delta t) = \phi_{\Delta t}(x(t))$$
$\phi_{\Delta t}$がデータから求められればデータの予測ができる。ガウス過程を関数$\phi_{\Delta t}$の事前分布にする方法を採用。
$$\phi_{\Delta t}　\sim \: \text{GP}(\mathbf{0},　\mathbf{K})$$

### $\phi_{\Delta t}$とヤコビ行列との関係
式(1)の力学系のヤコビ行列$\mathbf{J}$は$\mathbf{J} = \frac{\partial f}{\partial x}$で与えられる。これと$\phi_{\Delta t}$の間には次の関係式が成り立つ。
$$
\mathbf{J} \approx \frac{1}{\Delta t} \left(\frac{\partial \phi_{\Delta t}}{ \partial x} - \mathbf{I}\right)
$$
ただし、$\mathbf{I}$は恒等行列。データから$\phi_{\Delta t}$の偏導関数が求められれば力学系(1)のヤコビ行列が求められる。

### ガウス過程回帰による偏導関数の推定
ガウス過程回帰のパラメタをデータから学習した後の事後予測分布もガウス過程になる。
$$\phi_{\Delta t} \sim \: \text{GP}(\mu^* , \mathbf{K}^*)$$
$\mu^*$は関数$\phi_{\Delta t}$の事後期待値であり、$\mu^* = \text{E}[\phi_{\Delta t}]$について次の関係が成り立つので
$$
\begin{align}
\frac{\partial}{\partial x} \text{E}[\phi_{\Delta t}] = \text{E}[\frac{\partial \phi_{\Delta t}}{\partial x}] \notag
\end{align}
$$
$\phi_{\Delta t}$の偏導関数の事後期待値はガウス過程回帰の事後予測分布の期待値の微分から求められる。

- ガウス過程回帰のほうがバイアスが少なく、かつ精度が高く、ノイズに対しても頑健
- Deyle et al.(2016)のSIが述べるように、S-mapによる推定のバイアスの大部分は線形近似に由来するので、線形近似を行わず非線形関数を推定するガウス過程回帰のほうがバイスの少ない推定ができると考えられる
- ガウス過程回帰による推定はカーネルの選択に依存する
  - Maternカーネルの場合、ノイズがないか非常に弱い場合には極めて精度が高いが、ノイズが少しでも強いと精度が極めて落ちる
  - ガウシアンカーネルの場合、ノイズがない状況での精度はMaternカーネルより若干劣るが、ノイズがある状況でもより頑健な推定ができる
- ガウシアンカーネルの方が、高周波成分を除去してよりなめらかな関数を推定するのでノイズの影響を受けにくいのは理解できる
